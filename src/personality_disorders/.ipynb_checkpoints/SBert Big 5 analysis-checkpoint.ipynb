```python
from sentence_transformers import SentenceTransformer,util
model = SentenceTransformer('all-MiniLM-L6-v2').to("cuda:0")
import numpy as np
import pandas as pd
import torch
import dask
import dask.dataframe as dd
import re
from datetime import datetime
import csv
abs_path = '../../dataset' 

def tweetClean(tuit):
    tuit = str(tuit)
    tuit = re.sub('(www|https)[^\s]+', ' ', tuit)
    tuit = re.sub('#', ' ', tuit)
    tuit = re.sub('@', ' ', tuit)
    return tuit

def move_to(obj, device):
    if torch.is_tensor(obj):
        return obj.to(device)
    elif isinstance(obj, dict):
        res = {}
        for k, v in obj.items():
            res[k] = move_to(torch.tensor(v), device)
        return res
    else:
        raise TypeError("Invalid type for move_to")
        
```

```python
# Use this code block for Big5 (NOTE!!!: Run this cell OR the following one depending on the analysis you want to conduct)
dict_of_terms = {
    'E-POSITIVE' : ['dominant', 'assertive', 'authoritarian', 'forceful', 'assured', 'confident', 'firm', 'persistent'],
    'E-NEGATIVE' : ['nervous', 'modest', 'quiet', 'forceless', 'afraid', 'shy', 'calm', 'indecisive'],
    'A-POSITIVE' : ['Tender', 'Gentle', 'Soft', 'Kind', 'Affectionate', 'Helpful', 'Sympathetic', 'Friendly'],
    'A-NEGATIVE' : ['Cruel', 'Unfriendly', 'Negative', 'Mean', 'Brutal', 'Inconsiderate', 'Insensitive', 'Cold'],
    'C-POSITIVE' : ['Organized', 'Orderly', 'Tidy', 'Neat', 'Efficient', 'Persistent', 'Systematic', 'Straight', 'Careful', 'Reliable'],
    'C-NEGATIVE' : ['Distracted', 'Unreliable', 'Incompetent', 'Wild', 'Inefficient', 'Disloyal', 'Chaotic', 'Confused', 'Messy', 'Disorganized'],
    'N-POSITIVE' : ['Worried', 'Stressed', 'Anxious', 'Nervous', 'Fearful', 'Touchy', 'Guilty', 'Insecure', 'Restless', 'Emotional'],
    'N-NEGATIVE' : ['Balanced Stable', 'Confident', 'Fearless', 'Calm', 'Easy going', 'Relaxed', 'Secure', 'Comforted', 'Peaceful'],
    'O-POSITIVE' : ['Philosophical', 'Abstract', 'Imaginative', 'Curious', 'Reflective', 'Literary', 'Questioning', 'Individualistic', 'Unique', 'Open'],
    'O-NEGATIVE' : ['Narrow-minded', 'Concrete', 'Ordinary', 'Incurious', 'Thoughtless', 'Ignorant', 'Uneducated', 'Common', 'Conventional', 'Restricted']
}

dict_of_embeddings = {
    'E-POSITIVE' : None,
    'E-NEGATIVE' : None,
    'A-POSITIVE' : None,
    'A-NEGATIVE' : None,
    'C-POSITIVE' : None,
    'C-NEGATIVE' : None,
    'N-POSITIVE' : None,
    'N-NEGATIVE' : None,
    'O-POSITIVE' : None,
    'O-NEGATIVE' : None
}
```

```python
# Use this code block for disorders
dict_of_terms = {
    'Schizoid' : ['indifferent', 'apathetic', 'remote', 'solitary'],
    'Depressive' : ['sad', 'depressed', 'hopeless', 'gloomy', 'fatalistic'],
    'Avoidant' : ['shy', 'reflective', 'embarrassed', 'anxious'],
    'Dependent' : ['helpless', 'incapable', 'passive', 'immature'],
    'Histrionic' : ['dramatic', 'seductive', 'shallow', 'hyperactive', 'vain'],
    'Narcissistic' : ['selfish', 'arrogant', 'grandiose', 'indifferent'],
    'Compulsive' : ['restrained', 'conscientious', 'respectful', 'rigid'],
    'Paranoid' : ['cautious', 'defensive', 'distrustful', 'suspicious'],
    'Schizotypal' : ['eccentric', 'alien', 'bizarre', 'absent']
}

dict_of_embeddings = {
    'Schizoid' : None,
    'Depressive' : None,
    'Avoidant' : None,
    'Dependent' : None,
    'Histrionic' : None,
    'Narcissistic' : None,
    'Compulsive' : None,
    'Paranoid' : None,
    'Schizotypal' : None
}
```

```python
for x, y in dict_of_terms.items():
    dict_of_embeddings[x] = np.mean(model.encode(y),axis=0)
move_to(dict_of_embeddings, "cuda:0")
    
# Read CSVs and build up the dataframe
tweets_df1 = dd.read_csv(abs_path + '/jan_2021_US.csv',
                    lineterminator = '\n')
print(tweets_df1.head())
tweets=tweets_df1.loc[:, ['date', 'tweet']]
tweets=tweets.reset_index(drop=True) 
tweets = tweets.repartition(npartitions=100) 
tweets['tweet'] = tweets['tweet'].map(lambda x: tweetClean(x))
tweets['weekdate'] = dd.to_datetime(tweets['date'],format='%Y-%m-%d') - dd.to_datetime(tweets['date'],format='%Y-%m-%d').dt.weekday * np.timedelta64(1,'D')
```

```python
#Change file name to AnalysisBig5Us to store the Big5 analysis
xx = datetime.now()
print(xx)
with open(abs_path + 'AnalysisDisordersEs_sampled.csv', "a+") as csv_file: #Renamed from AnalysisBig5Usa to AnalysisDisorders
    writer = csv.writer(csv_file, delimiter=',')
    writer.writerow(['weekdate','tweet']+list(dict_of_embeddings.keys()))
    i=0
    for df_chunk in tweets.partitions:
        output = (model.encode(list(df_chunk['tweet'].compute()),convert_to_tensor=True).to("cuda:0"))
        for tweet, weekdate, emb in zip(df_chunk.tweet.compute(),df_chunk.weekdate.compute(),output):
            l = []
            for x in dict_of_embeddings.keys():
                l.append(util.cos_sim(emb,dict_of_embeddings.get(x)).item())
            writer.writerow([weekdate,tweet]+l)
        i+=1
        #print(i*100000) 
csv_file.close()
print("Elapsed time: ")
print(datetime.now()-xx)            
```

```python
# Code block for grouping disorders data.

import pandas as pd
import csv
import json
import re

collist = list(dict_of_embeddings.keys())
collist.insert(0,'weekdate')

disorders_df = pd.read_csv(abs_path + 'AnalysisDisordersEs_sampled.csv',usecols=collist)
disorders_df = disorders_df.groupby(['weekdate'])[list(dict_of_embeddings.keys())].mean().reset_index()
disorders_df.to_csv(abs_path + 'disorders_groupedEs_sampled.csv',header=True, index=False)
```
